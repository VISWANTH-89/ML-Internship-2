```python
from sklearn.linear_model import LinearRegression

# Select one feature for visualization
X_lr = dia[['plas']]
y_lr = dia['class'].map({'tested_positive': 1, 'tested_negative': 0})

# Train Linear Regression
lin_reg = LinearRegression()
lin_reg.fit(X_lr, y_lr)

# Predictions
y_lr_pred = lin_reg.predict(X_lr)

# Plot
plt.figure(figsize=(8,5))
plt.scatter(X_lr, y_lr, alpha=0.5, label='Actual Data')
plt.plot(X_lr, y_lr_pred, color='red', linewidth=2, label='Linear Regression Line')

plt.xlabel('Plasma Glucose')
plt.ylabel('Diabetes Class (0: Negative, 1: Positive)')
plt.title('Linear Regression Line: Plasma vs Diabetes')
plt.legend()
plt.show()
```
