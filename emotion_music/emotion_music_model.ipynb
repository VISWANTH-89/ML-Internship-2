```python
from google.colab import files
from gtts import gTTS
from pydub import AudioSegment
import IPython.display as ipd
import docx # Import the python-docx library

# Upload text file
uploaded = files.upload()
text_file = list(uploaded.keys())[0]

# Check file extension and read content accordingly
if text_file.endswith(".docx"):
    try:
        document = docx.Document(text_file)
        text = "\n".join([paragraph.text for paragraph in document.paragraphs])
    except Exception as e:
        print(f"Error reading DOCX file: {e}")
        text = "" # Set text to empty string in case of error
else:
    # Assume it's a plain text file
    with open(text_file, "r") as f:
        text = f.read()

# Let user choose emotion
print("Choose emotion for background music:")
print("1 - Happy")
print("2 - Sad")
print("3 - Calm")
print("4 - Energetic")

emotion_choice = int(input("Enter number: "))

# Map emotion to background music file
emotion_to_music = {
    1: "happy.mp3",
    2: "sad.mp3",
    3: "calm.mp3",
    4: "energetic.mp3"
}

bg_music_file = emotion_to_music.get(emotion_choice, "calm.mp3")
```

```python
import pickle
from google.colab import files

# Define emotion to background music mapping
emotion_music_model = {
    "emotions": {
        "happy": "happy.mp3",
        "sad": "sad.mp3",
        "calm": "calm.mp3",
        "energetic": "energetic.mp3"
    }
}

# Save as PKL file
pkl_filename = "emotion_music_model.pkl"
with open(pkl_filename, "wb") as f:
    pickle.dump(emotion_music_model, f)

print(f"✅ PKL file saved as: {pkl_filename}")

# Download PKL file
files.download(pkl_filename)
```
