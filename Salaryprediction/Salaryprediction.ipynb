```python
#Import libraries
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
#load data
saldf=pd.read_csv('/content/Salary Data.csv')
```

```python
#view data
saldf.head()
```

```python
saldf.info()
```

```python
saldf.isnull().sum()
```

```python
#prepare independent and dependent variables
inp=saldf[['Years of Experience']]
out=saldf['Salary']
```

```python
#create an instance of LR
LR=LinearRegression()
```

```python
#train the model
df_cleaned = saldf.dropna(subset=['Years of Experience', 'Salary'])

inp_cleaned = df_cleaned[['Years of Experience']]
out_cleaned = df_cleaned['Salary']

LR.fit(inp_cleaned, out_cleaned)
```

```python
#predict when experience is 5years
LR.predict([[5]])
```

```python
plt.scatter(inp,out, color='blue', label='Actual Data')
plt.plot(inp,out, color='red', linewidth=2, label='Regression Line')

plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Linear Regression Plot')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pickle
with open("Salaryprediction.pkl","wb") as f:
  pickle.dump(LR,f)
from google.colab import files
files.download('Salaryprediction.pkl')
```
